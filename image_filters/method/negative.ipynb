```python
import cv2
import matplotlib.pyplot as plt
```

```python
#read image from storage
img = cv2.imread('/home/scyeroux/python/image_processing-basics/images/Lenna.png')
#convert to RGB channel
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img,)
```

```python
histr = cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(histr)
plt.show()
```

```python
#function to convert to negative image using open cv
def negative_using_openCV(img):
    return cv2.bitwise_not(img)
```

```python
cv_negative_image = negative_using_openCV(img)
plt.imshow(cv_negative_image)
```

```python
histr = cv2.calcHist([cv_negative_image],[0],None,[256],[0,256])
plt.plot(histr)
plt.show()
```
