```python
import cv2
import matplotlib.pyplot as plt
```

```python
#read image from storage
img = cv2.imread('/home/scyeroux/python/image_processing-basics/images/Lenna.png')
#convert to RGB channel
img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
plt.imshow(img,)
```

```python
histr = cv2.calcHist([img],[0],None,[256],[0,256])
plt.plot(histr)
plt.show()
```

```python
# Obtain the dimensions of the image array
# using the shape method
(row, col) = img.shape[0:2]
 
# Take the average of pixel values of the BGR Channels
# to convert the colored image to grayscale image
for i in range(row):
    for j in range(col):
        # Find the average of the BGR pixel values
        img[i, j] = sum(img[i, j]) * 0.33
```

```python
img_grascale = plt.imshow(img)
```
